```python
# importing libraries

from PIL import Image
import matplotlib.pyplot as plt
```

```python
def pixelifier (image, small_size):
    
    #reading the image
    img = Image.open(image)
    
    #Grabbing the dimensions
    orig_size = img.size 
    
    #Sizing down the image
    tiny_image = img.resize(small_size, Image.BILINEAR)
    #Resizing our image to its original size
    out_size = tiny_image.resize(orig_size, Image.NEAREST)
    
    #Saving our pixelated image
    out_size.save(f'Image_{small_size[0]}X{small_size[1]}.png')
    
    #Displaying our images
    plt.figure(figsize=(16,10))
    #original image
    plt.subplot(1,2,1)
    plt.title(f'Original image {img.size[0]}x{img.size[1]}', size=18)
    plt.imshow(img)   
    plt.axis('off')   
    #pixel art
    plt.subplot(1,2,2)
    plt.title(f'Image {small_size[0]}x{small_size[1]}', size=18)
    plt.imshow(out_size)
    plt.axis('off')
    plt.show()
    
    
    
```

```python
pixelifier('kimetsu_pixelation_time.jpg', (120,80))
```
